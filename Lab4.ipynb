**N-Queens Problem**

```python
def is_safe(board, row, column, size):
    for previous_row in range(row):
        if board[previous_row] == column:
            return False
        if abs(board[previous_row] - column) == abs(previous_row - row):
            return False
    return True


def solve_n_queens(board, row, size, solutions):
    if row == size:
        solutions.append(board[:])
        return

    for column in range(size):
        if is_safe(board, row, column, size):
            board[row] = column
            solve_n_queens(board, row + 1, size, solutions)


def display_solutions(solutions, size):
    for index, solution in enumerate(solutions, start=1):
        print("Solution", index)
        for row in solution:
            line = ""
            for column in range(size):
                if column == row:
                    line += "Q "
                else:
                    line += ". "
            print(line)
        print()


def main():
    size = int(input("Enter the number of queens: "))
    board = [-1] * size
    solutions = []

    solve_n_queens(board, 0, size, solutions)

    if len(solutions) == 0:
        print("No solutions found")
    else:
        print("Total solutions:", len(solutions))
        display_solutions(solutions, size)


main()
```

**Longest Common Subsequence (LCS)**

```python
def lcs_backtracking(s1, s2, i, j):
    if i == 0 or j == 0:
        return 0

    if s1[i - 1] == s2[j - 1]:
        return 1 + lcs_backtracking(s1, s2, i - 1, j - 1)
    else:
        return max(
            lcs_backtracking(s1, s2, i - 1, j),
            lcs_backtracking(s1, s2, i, j - 1)
        )


s1 = input("Enter the first string: ")
s2 = input("Enter the second string: ")

length = lcs_backtracking(s1, s2, len(s1), len(s2))
print("The length of the Longest Common Subsequence is:", length)
```

**Subset Sum Problem**

```python
def find_subsets(numbers, target, index, current, current_sum, result):
    if current_sum == target:
        result.append(current[:])
        return

    if index == len(numbers) or current_sum > target:
        return

    current.append(numbers[index])
    find_subsets(numbers, target, index + 1, current, current_sum + numbers[index], result)
    current.pop()

    find_subsets(numbers, target, index + 1, current, current_sum, result)


def main():
    numbers = list(map(int, input("Enter the elements: ").split()))
    target = int(input("Enter the target sum: "))

    result = []
    find_subsets(numbers, target, 0, [], 0, result)

    if len(result) == 0:
        print("No subset found")
    else:
        print("Subsets with given sum:")
        for subset in result:
            print(subset)


main()
```
