Assignment 1

```python
memo = {}

def fibonacci(n):
    if n in memo:
        return memo[n]
    if n == 0:
        return 0
    if n == 1:
        return 1
    result = fibonacci(n-1) + fibonacci(n-2)
    memo[n] = result
    return result

n = int(input("Enter the number: "))
print("Fibonacci of", n, "is:", fibonacci(n))
```

Assignment 2

```python
first = input("Enter number of packages and capacity separated by space: ")
n, capacity = map(int, first.split())

values = []
weights = []

for i in range(n):
    line = input("Enter value and weight of package " + str(i+1) + ": ")
    v, w = map(int, line.split())
    values.append(v)
    weights.append(w)

dp = [[0 for i in range(capacity + 1)] for j in range(n + 1)]

for i in range(1, n + 1):
    for w in range(1, capacity + 1):
        if weights[i-1] <= w:
            dp[i][w] = max(values[i-1] + dp[i-1][w - weights[i-1]], dp[i-1][w])
        else:
            dp[i][w] = dp[i-1][w]


print("The maximum value the drone can carry is:", dp[n][capacity])
```

Assignment 3

```python
s1 = input("Enter the first string: ")
s2 = input("Enter the second string: ")

m = len(s1)
n = len(s2)

dp = [[0 for i in range(n + 1)] for j in range(m + 1)]

for i in range(1, m + 1):
    for j in range(1, n + 1):
        if s1[i-1] == s2[j-1]:
            dp[i][j] = dp[i-1][j-1] + 1
        else:
            if dp[i-1][j] > dp[i][j-1]:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = dp[i][j-1]

print("The length of the Longest Common Subsequence is:", dp[m][n])
```

Assignment 4

```python
n = int(input("Enter number of different coins: "))

coins = []
for i in range(n):
    c = int(input("Enter value of coin " + str(i+1) + ": "))
    coins.append(c)

amount = int(input("Enter the amount to return: "))

dp = [1000000] * (amount + 1)
dp[0] = 0

for i in range(1, amount + 1):
    for c in coins:
        if c <= i:
            if dp[i - c] + 1 < dp[i]:
                dp[i] = dp[i - c] + 1

if dp[amount] == 1000000:
    print("It is not possible to make the amount with given coins")
else:
    print("Minimum number of coins needed is:", dp[amount])
```

Assignment 5

```python
n = int(input("Enter number of matrices: "))

print("Enter the dimensions in one line separated by space:")
dims = list(map(int, input().split()))

dp = [[0 for i in range(n)] for j in range(n)]

for length in range(2, n):
    for i in range(0, n - length):
        j = i + length
        dp[i][j] = 10**18
        for k in range(i, j):
            cost = dp[i][k] + dp[k][j] + dims[i] * dims[k] * dims[j]
            if cost < dp[i][j]:
                dp[i][j] = cost

print("Minimum number of scalar multiplications needed is:", dp[0][n-1])
```

```python
num = int(input("Enter a number: "))

result = ""

while num > 0:
    num -= 1
    remainder = num % 26
    result = chr(ord('a') + remainder) + result
    num = num // 26

print("Output:", result)
```
