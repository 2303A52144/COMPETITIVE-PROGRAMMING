```python
def solve():
    T = int(input())
    results = []

    for _ in range(T):
        N = int(input())
        meetings = []
        for _ in range(N):
            s, e = map(int, input().split())
            meetings.append((s, e))

        # Sort meetings by their end times
        meetings.sort(key=lambda x: x[1])
        count = 0
        last_end_time = -1  # Initialize with a time that allows the first meeting to be picked

        for s, e in meetings:
            # If the current meeting's start time is after or at the same time as the last selected meeting's end time
            if s >= last_end_time:
                count += 1
                last_end_time = e
        results.append(count)

    for res in results:
        print(res)

solve()
```

### Shelf Product Lookup (Divide and Conquer with Binary Search)

The problem requires us to count the occurrences of a given product ID `X` in a sorted list of product IDs. The constraint is to use a binary-search style divide and conquer approach, specifically by finding the first and last occurrences of `X`.

**Algorithm:**
1.  **`find_first_occurrence(arr, x)` Function:**
    *   This binary search variant aims to find the *leftmost* index where `x` appears in `arr`.
    *   When `arr[mid] == x`, we record `mid` as a potential first occurrence and continue searching in the left half (`high = mid - 1`) to see if an even earlier `x` exists.
    *   If `x` is not found, it returns -1.

2.  **`find_last_occurrence(arr, x)` Function:**
    *   This binary search variant aims to find the *rightmost* index where `x` appears in `arr`.
    *   When `arr[mid] == x`, we record `mid` as a potential last occurrence and continue searching in the right half (`low = mid + 1`) to see if an even later `x` exists.
    *   If `x` is not found, it returns -1.

3.  **Main `solve()` Function:**
    *   Reads the number of test cases `T`.
    *   For each test case:
        *   Reads `N` (number of products) and `Q` (number of queries).
        *   Reads the `N` sorted product IDs into a list `arr`.
        *   For each of the `Q` queries:
            *   Reads the query ID `X`.
            *   Calls `find_first_occurrence(arr, X)` to get the start index.
            *   Calls `find_last_occurrence(arr, X)` to get the end index.
            *   If `first_idx` is -1 (meaning `X` was not found in the list), the count is 0.
            *   Otherwise, the count is `last_idx - first_idx + 1`.
            *   Prints the count for the current query.

```python
def find_first_occurrence(arr, x):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = low + (high - low) // 2
        if arr[mid] == x:
            result = mid  # Found a potential first occurrence
            high = mid - 1  # Try to find an earlier one in the left half
        elif arr[mid] < x:
            low = mid + 1  # x must be in the right half
        else:
            high = mid - 1  # x must be in the left half
    return result

def find_last_occurrence(arr, x):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = low + (high - low) // 2
        if arr[mid] == x:
            result = mid  # Found a potential last occurrence
            low = mid + 1  # Try to find a later one in the right half
        elif arr[mid] < x:
            low = mid + 1  # x must be in the right half
        else:
            high = mid - 1  # x must be in the left half
    return result

def solve():
    T = int(input())
    for _ in range(T):
        N, Q = map(int, input().split())
        arr = list(map(int, input().split()))

        for _ in range(Q):
            x = int(input())

            first_idx = find_first_occurrence(arr, x)
            last_idx = find_last_occurrence(arr, x)

            if first_idx == -1:
                print(0)
            else:
                print(last_idx - first_idx + 1)

solve()
```

### Shelf Product Lookup (Divide and Conquer with Binary Search)

The problem requires us to count the occurrences of a given product ID `X` in a sorted list of product IDs. The constraint is to use a binary-search style divide and conquer approach, specifically by finding the first and last occurrences of `X`.

**Algorithm:**
1.  **`find_first_occurrence(arr, x)` Function:**
    *   This binary search variant aims to find the *leftmost* index where `x` appears in `arr`.
    *   When `arr[mid] == x`, we record `mid` as a potential first occurrence and continue searching in the left half (`high = mid - 1`) to see if an even earlier `x` exists.
    *   If `x` is not found, it returns -1.

2.  **`find_last_occurrence(arr, x)` Function:**
    *   This binary search variant aims to find the *rightmost* index where `x` appears in `arr`.
    *   When `arr[mid] == x`, we record `mid` as a potential last occurrence and continue searching in the right half (`low = mid + 1`) to see if an even later `x` exists.
    *   If `x` is not found, it returns -1.

3.  **Main `solve()` Function:**
    *   Reads the number of test cases `T`.
    *   For each test case:
        *   Reads `N` (number of products) and `Q` (number of queries).
        *   Reads the `N` sorted product IDs into a list `arr`.
        *   For each of the `Q` queries:
            *   Reads the query ID `X`.
            *   Calls `find_first_occurrence(arr, X)` to get the start index.
            *   Calls `find_last_occurrence(arr, X)` to get the end index.
            *   If `first_idx` is -1 (meaning `X` was not found in the list), the count is 0.
            *   Otherwise, the count is `last_idx - first_idx + 1`.
            *   Prints the count for the current query.

```python
def first_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            high = mid - 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans

def last_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            low = mid + 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans


# ---- MAIN PROGRAM ----
T = int(input())

for _ in range(T):
    N, Q = map(int, input().split())
    arr = list(map(int, input().split()))

    for _ in range(Q):
        x = int(input())
        first = first_occurrence(arr, x)
        if first == -1:
            print(0)
        else:
            last = last_occurrence(arr, x)
            print(last - first + 1)

```
